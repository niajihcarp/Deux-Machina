```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 100, 100)
y = x * 2


fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")



ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.plot(x, y)
ax2.set_xlabel("x")
ax2.set_ylabel("y")



plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(0,100)
y = x*2
z = x**2

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, z, 'b')



ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, y, 'b')



ax2.set_xlim(20, 22)
ax2.set_ylim(40, 48)

plt.show()
```
