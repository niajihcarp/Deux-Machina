## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt

# data
x = [0, 100]
y = [0, 200]

# create figure
fig = plt.figure()

# add axes at [left, bottom, width, height]
ax = fig.add_axes([0, 0, 1, 1])

# plot
ax.plot(x, y)

# labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

plt.show()
```

**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
import matplotlib.pyplot as plt

# data
x = [0, 100]
y = [0, 200]

# create figure
fig = plt.figure()

# add first axes
ax1 = fig.add_axes([0, 0, 1, 1])

# add second axes
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

# plot on both axes
ax1.plot(x, y)
ax2.plot(x, y)

# show figure
plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np

# data
x = np.linspace(0, 100, 100)
z = x**2

# create figure
fig = plt.figure()

# main axes
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('zoom')

# inset (zoomed) axes
ax2 = fig.add_axes([0.2, 0.55, 0.3, 0.3])
ax2.plot(x, z)
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt

# data
x = [0, 100]
y = [0, 200]
z = [100, 0]

# create subplots (1 row, 2 columns)
fig, axes = plt.subplots(nrows=1, ncols=2)

# first plot: (x, y)
axes[0].plot(x, y, linewidth=2, linestyle='-')
axes[0].set_title('x vs y')

# second plot: (x, z)
axes[1].plot(x, z, linewidth=2, linestyle='--')
axes[1].set_title('x vs z')

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np

# data
x = np.linspace(0, 100, 100)
y = 2 * x
z = x**2

# create subplots WITH figsize
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# left plot: x vs y
axes[0].plot(x, y, 'b--')
axes[0].set_title('Linear')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# right plot: x vs z
axes[1].plot(x, z, 'r')
axes[1].set_title('Quadratic')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

plt.show()
```
